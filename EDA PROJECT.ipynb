# 8. Which day of the week has maximum sales? 

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('downloads/supermarket sales Sheet1.csv')
df['Date'] = pd.to_datetime(df['Date'])
df['weekday']=df['Date'].dt.day_name()
plt.figure(figsize=(8, 6))
plt.title('Sales as Per the Week Day')
plt.grid(True)
sns.countplot(df['weekday'])

```

```python
# here we want to track the day in which the max sales happened, i plotted it with count plot i convet the date with weekday and plotted it
# here we can see that the on saturday maximun sales happened might be this is the week end day so this happened after that max
# sales happened on tuesday.
```

# 9. Which hour of the day is busiest? 

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('downloads/supermarket sales Sheet1.csv')
df['Time'] = pd.to_datetime(df['Time'])
df['Hours']=df['Time'].dt.hour
plt.figure(figsize=(8, 6))
plt.title('quantity sold per hours in the day')
plt.grid(True)
sns.lineplot(x=df['Hours'],y=df['Quantity'], data=df)
```

```python
# i chosen to plot the line plot between time and quantity because line graph gives liner relation (decreasing/increasing) 
#between two variable if they related to each other, here from graph we can say that the(14 hr) i.e 2 pm is the most busiest time 
#or at 2 pm highest quantity of product sold
```
